* 코드 자동완성, 챗봇기능, Chat GPT와 동일한 방식으로 사용 가능
* Colab 내에서 언제든지 활용 가능
* 원래는 코랩 프로 유료 구독자들에게만 제공되었으나 얼마전에 무료로 기능을 사용할 수 있게 되었습니다.

```python
import random

# Create a 5x5 grid of numbers
grid = [[0 for i in range(5)] for j in range(5)]

# Fill the grid with random numbers from 1 to 25
for i in range(5):
  for j in range(5):
    grid[i][j] = random.randint(1, 25)

# Choose a random number to be the "bingo" number
bingo_number = random.randint(1, 25)

# Print the grid
for row in grid:
  print(' '.join(map(str, row)))

# Get the user's input
user_input = input('Enter a number: ')

# Check if the user's input is the bingo number
if user_input == bingo_number:
  print('You win!')
else:
  print('You lose.')
```
